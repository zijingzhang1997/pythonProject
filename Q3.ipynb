```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,1,1000)


def function(K):
    f=2
    t = np.linspace(0,1,1000)
    pi=np.pi
    y=np.array
    for j in range(len(t)):
        y=0
        for i in range(1,K+1):
            s=np.sin(2*pi*(2*i-1)*f*t)
            sum =s/(2*i-1)
            y +=sum
    return y
y=function(1)
y1=y.copy()

```

```python
plt.plot(t,y1)
plt.xlabel('t')
plt.ylabel('y1')
plt.title('y1 curve')
```

```python
y=function(3)
y2=y.copy()
plt.plot(t,y2)
plt.xlabel('t')
plt.ylabel('y2')
plt.title('y2 curve')
```

```python
y=function(9)
y3=y.copy()
plt.plot(t,y3)
plt.xlabel('t')
plt.ylabel('y3')
plt.title('y3 curve')







```

```python
y=function(50)
y4=y.copy()
plt.plot(t,y4)
plt.xlabel('t')
plt.ylabel('y4')
plt.title('y4 curve')
```

```python
y=function(1000000)
y5=y.copy()
plt.plot(t,y5)
plt.xlabel('t')
plt.ylabel('y5')
plt.title('y5 curve')
```

```python
fig = plt.figure()


# Each plot has unique axes

ax1 = plt.subplot(321) # equivalent to plt.subplot(2, 2, 1)
ax2 = plt.subplot(322)
ax3 = plt.subplot(323)
ax4 = plt.subplot(324)
ax5 = plt.subplot(325)

ax1.plot(t,y1)
ax2.plot(t,y2)
ax3.plot(t,y3)
ax4.plot(t,y4)
ax5.plot(t,y5)

# Can add labels and title to each individual subplot
ax1.set_xlabel('t')
ax1.set_ylabel('y1')

ax2.set_xlabel('t')
ax2.set_ylabel('y2')

ax3.set_xlabel('t')
ax3.set_ylabel('y3')

ax4.set_xlabel('t')
ax4.set_ylabel('y4')

ax4.set_xlabel('t')
ax5.set_ylabel('y5')
plt.suptitle('curve') #Setting the title for the whole plot

plt.tight_layout()
```
