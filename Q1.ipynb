```python
x=[3,4,1,2,5,6,7,8,2,4]   #1.(a) answer    first transfer h(n) to h(-n) then add up x(n)*h(k-n) fro all different n  then add up k to see the final result
h=[0.25,0.25,0.25]
M=len(x)
N=len(h)

h_flip=[]
for i in range(N):
    h_flip.append(h[N-1-i])
print(h_flip)

x_large=[]
for i in range(N-1):
    x_large.append(0)
for i in range(M):
    x_large.append(x[i])
for i in range(N-1):
    x_large.append(0)
print(x_large)

conv=[]
for i in range(M+N-1):
    summary=0
    for j in range(N):
        summary +=h_flip[j]*x_large[i+j]
    conv.append(summary)
print(conv)
```

```python
import numpy as np     #1.(d)
import time
start = time.clock()

 
a = np.array([3,4,1,2,5,6,7,8,2,4])
x = np.tile(a, (100,1)) 
x.shape=1000,
b=[0.25,0.25,0.25]
h = np.tile(b, (100,1)) 
h.shape=300,
   

M=len(x)
N=len(h)
print(M)
print(N)
#print(x)
#print(h)
h_flip=[]
for i in range(N):
    h_flip.append(h[N-1-i])


x_large=[]
for i in range(N-1):
    x_large.append(0)
for i in range(M):
    x_large.append(x[i])
for i in range(N-1):
    x_large.append(0)


conv=[]
for i in range(M+N-1):
    summary=0
    for j in range(N):
        summary +=h_flip[j]*x_large[i+j]
    conv.append(summary)
#print(conv)

end = time.clock()

t=end-start
print(conv)
print("conventional Runtime is ：",t)


start = time.clock()
y=np.zeros(N+M-1)
for i in range(N):
    y[i]=np.sum(x[0:i+1])
for i in range(N,M-1):
    y[i]=np.sum(x[i-N:i])
for i in range(M-1,N+M-1):
    y[i]=np.sum(x[i-N+1:N])
y=0.25*y    
end = time.clock()

t=end-start

print("efficient Runtime is ：",t)
```
