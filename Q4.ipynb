```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
t=np.arange(1000)
def function(b):
        h1 = signal.unit_impulse(1000, 2)*b
        n = np.arange(1000)
        x = np.sin(0.1*n)
        M=len(x)
        N=len(h1)

        h_flip=[]
        for i in range(N):
            h_flip.append(h1[N-1-i])


        x_large=[]
        for i in range(N-1):
            x_large.append(0)
        for i in range(M):
            x_large.append(x[i])
        for i in range(N-1):
            x_large.append(0)


        y1=[]
        for i in range(M+N-1):
            summary=0
            for j in range(N):
                summary +=h_flip[j]*x_large[i+j]
            y1.append(summary)

        t=np.arange(1000)
       


        h2=np.zeros(1000)   
        for i in range(200):
            h2[i]=-1

        h2_flip=[]
        for i in range(N):
            h2_flip.append(h2[N-1-i])


        y1_large=[]
        for i in range(N-1):
            y1_large.append(0)
        for i in range(M):
            y1_large.append(y1[i])
        for i in range(N-1):
            y1_large.append(0)


        y2=[]
        for i in range(M+N-1):
            summary=0
            for j in range(N):
                summary +=h2_flip[j]*y1_large[i+j]
            y2.append(summary)
        
        return y1,y2


(y1,y2)=function(-5)
(y3,y4)=function(-1)
(y5,y6)=function(5)

fig = plt.figure()

ax1 = plt.subplot(321
ax2 = plt.subplot(322)
ax3 = plt.subplot(323)
ax4 = plt.subplot(324)
ax5 = plt.subplot(325)
ax6 = plt.subplot(326)

ax1.plot(t,y1[0:1000])
ax2.plot(t,y2[0:1000])
ax3.plot(t,y3[0:1000])
ax4.plot(t,y4[0:1000])
ax5.plot(t,y5[0:1000])
ax6.plot(t,y6[0:1000])
fig.tight_layout() 
                  
# What do you observe abouty1 and how it relates to b and x. How do you explain this behavior? How is y2 related to y1 and x?
# b changes the max and min of y1. y1 has the same shape with x, but with a phase shift due to (n-2). y1 is the impulse response of x
# y2 is the unit step response difference between u[n-200] and u[n] of y1, and since y1 is impluse response of x, y2 also has the shape of the unit step response  difference between u[n-200] and u[n] of x
```
