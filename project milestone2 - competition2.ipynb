```python
#1.1 
T0=np.array([1,1,0,0,0])#T=np.array([sx, thetax,sy,thetay, tx,ty])



def transform(img_m,T,img_f_size):
    xmax=img_m.shape[0]
    ymax=img_m.shape[1]

    x0=np.linspace(-(xmax/2) +1,xmax/2,xmax)   #x,y[-127,128]
    y0=np.linspace(-(ymax/2) +1,ymax/2,ymax)
    x,y=np.meshgrid(x0,y0)
    x_t=np.zeros([xmax,ymax])
    y_t=np.zeros([xmax,ymax])
    
    Trans=([T[0]*np.cos(T[2]),-T[0]*np.sin(T[2]),T[3]],[T[1]*np.sin(T[2]),T[1]*np.cos(T[2]),T[4]],[0,0,1])
    for i in range(xmax):
        for j in range(ymax):
          
            index=np.array([x[i,j],y[i,j],1])
            T_index=np.dot(Trans,index)
            T_index=np.array(T_index)
            x_t[i,j]=T_index[0]
            y_t[i,j]=T_index[1]
   
    
    x0f=np.linspace(-(img_f_size[0]/2)+1,img_f_size[0]/2,img_f_size[0])  #x,y[-127,128]
    y0f=np.linspace(-(img_f_size[1]/2)+1,img_f_size[1]/2,img_f_size[1])
    xf,yf=np.meshgrid(x0f,y0f)
    
    img_moved = griddata((x_t.flatten(),y_t.flatten()), img_m.flatten(), (xf, yf), method='linear',fill_value=0)#resample

    return img_moved


#1.2
def lossFunction(T,img_m,img_f):
    img_moved=transform(img_m,T,img_f.shape)
    #L=np.sum(math.pow(img_f[:,:] -img_moved[:,:],2))
    L=0
    for i in range(img_f.shape[0]):
        for j in range(img_f.shape[1]):
            L+=(img_f[i,j] -img_moved[i,j])**2

    return L



# res_t01c08 = minimize(lossFunction, T0,args=(img_t01,img_c08),method='Powell',options={'maxiter': 50})
# #res1 = minimize(lossFunction, T,args=(img1,img3),method='Nelder-Mead',options={'maxiter': 100})
# print(res_t01c08)
# Topt_t01c08=res_t01c08.x
# np.save(r'C:\Users\Edwin Kan Group\Documents\Zijing\project\Toptc\Topt_t01c08.npy',Topt_t01c08)
# img_moved_t01c08=transform(img_t01,Topt_t01c08,img_c08.shape)
# fig = plt.figure()
# plt.imshow(img_t01)
# plt.title('training01')
# fig = plt.figure()
# plt.imshow(img_c08) 
# plt.title('validation07')
# fig = plt.figure()
# plt.imshow(img_moved_t01c08)
# plt.title('training01 moved ')



# res_t01c14 = minimize(lossFunction, T0,args=(img_t01,img_c14),method='Powell',options={'maxiter': 50})
# Topt_t01c14=res_t01c14.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t01c14.npy',Topt_t01c14)
# res_t02c14 = minimize(lossFunction, T0,args=(img_t02,img_c14),method='Powell',options={'maxiter': 50})
# Topt_t02c14=res_t02c14.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t02c14.npy',Topt_t02c14)
# res_t03c14 = minimize(lossFunction, T0,args=(img_t03,img_c14),method='Powell',options={'maxiter': 50})
# Topt_t03c14=res_t03c14.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t03c14.npy',Topt_t03c14)
# res_t04c14 = minimize(lossFunction, T0,args=(img_t04,img_c14),method='Powell',options={'maxiter': 50})
# Topt_t04c14=res_t04c14.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t04c14.npy',Topt_t04c14)
# res_t05c14 = minimize(lossFunction, T0,args=(img_t05,img_c14),method='Powell',options={'maxiter': 50})
# Topt_t05c14=res_t05c14.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t05c14.npy',Topt_t05c14)
# res_t06c14= minimize(lossFunction, T0,args=(img_t06,img_c14),method='Powell',options={'maxiter': 50})
# Topt_t06c14=res_t06c14.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t06c14.npy',Topt_t06c14)

# res_t01c16 = minimize(lossFunction, T0,args=(img_t01,img_c16),method='Powell',options={'maxiter': 50})
# Topt_t01c16=res_t01c16.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t01c16.npy',Topt_t01c16)
# res_t02c16 = minimize(lossFunction, T0,args=(img_t02,img_c16),method='Powell',options={'maxiter': 50})
# Topt_t02c16=res_t02c16.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t02c16.npy',Topt_t02c16)
# res_t03c16 = minimize(lossFunction, T0,args=(img_t03,img_c16),method='Powell',options={'maxiter': 50})
# Topt_t03c16=res_t03c16.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t03c16.npy',Topt_t03c16)
# res_t04c16 = minimize(lossFunction, T0,args=(img_t04,img_c16),method='Powell',options={'maxiter': 50})
# Topt_t04c16=res_t04c16.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t04c16.npy',Topt_t04c16)
# res_t05c16 = minimize(lossFunction, T0,args=(img_t05,img_c16),method='Powell',options={'maxiter': 50})
# Topt_t05c16=res_t05c16.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t05c16.npy',Topt_t05c16)
# res_t06c16= minimize(lossFunction, T0,args=(img_t06,img_c16),method='Powell',options={'maxiter': 50})
# Topt_t06c16=res_t06c16.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t06c16.npy',Topt_t06c16)

# res_t01c17 = minimize(lossFunction, T0,args=(img_t01,img_c17),method='Powell',options={'maxiter': 50})
# Topt_t01c17=res_t01c17.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t01c17.npy',Topt_t01c17)
# res_t02c17 = minimize(lossFunction, T0,args=(img_t02,img_c17),method='Powell',options={'maxiter': 50})
# Topt_t02c17=res_t02c17.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t02c17.npy',Topt_t02c17)
# res_t03c17 = minimize(lossFunction, T0,args=(img_t03,img_c17),method='Powell',options={'maxiter': 50})
# Topt_t03c17=res_t03c17.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t03c17.npy',Topt_t03c17)
# res_t04c17 = minimize(lossFunction, T0,args=(img_t04,img_c17),method='Powell',options={'maxiter': 50})
# Topt_t04c17=res_t04c17.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t04c17.npy',Topt_t04c17)
# res_t05c17 = minimize(lossFunction, T0,args=(img_t05,img_c17),method='Powell',options={'maxiter': 50})
# Topt_t05c17=res_t05c17.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t05c17.npy',Topt_t05c17)
# res_t06c17= minimize(lossFunction, T0,args=(img_t06,img_c17),method='Powell',options={'maxiter': 50})
# Topt_t06c17=res_t06c17.x
# np.save(r'F:\lecture document\ECE 4250 signal\project\Toptc\Topt_t06c17.npy',Topt_t06c17)
            
                
                
```
